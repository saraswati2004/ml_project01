```python
x = df.drop(columns=['salary_in_usd'])
y = df['salary_in_usd']
x =pd.get_dummies(x)
x
```

```python
df.shape
```

```python
# train
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
se = StandardScaler()
```

```python
X_train = se.fit_transform(X_train)
X_test = se.transform(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,root_mean_squared_error
```

```python
def evaluate(model,X_train,y_train ,X_test,y_test):
  train_pred = model.predict(X_train)
  test_pred = model.predict(X_test)
  print(f"Accuracy for your {model} is")
  print(f'training r2 score for {model} is {r2_score(y_train,train_pred)}')
  print(f'testing r2 score for {model} is {r2_score(y_test,test_pred)}')

  # MAE
  print(f'training MAE for {model} is {mean_absolute_error(y_train,train_pred)}')
  print(f'testing MAE for {model} is {mean_absolute_error(y_test,test_pred)}')

  # RMSE
  print(f'training RMSE for {model} is {root_mean_squared_error(y_train,train_pred)}')
  print(f'testing RSME for {model} is {root_mean_squared_error(y_test,test_pred)}')
```

```python

from sklearn.linear_model import LinearRegression,Ridge
LR_model = LinearRegression()
LR_model.fit(X_train,y_train)
```

```python
evaluate(LR_model,X_train,y_train,X_test,y_test)
```

```python
y_pred = LR_model.predict(X_test)
```

```python
y_pred_train = LR_model.predict(X_train)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)#test data accuracy
r2_score(y_train,y_pred_train) #train data accuracy
```

```python
evaluate(best_ridge,X_train,y_train,X_test,y_test)
```
